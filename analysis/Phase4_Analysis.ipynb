```python
import pandas as pd

df = pd.read_csv("cyclistic_cleaned.csv")
df.head()
```

```python
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
```

```python
# If ride_length is a string like '0 days 00:30:16', first convert to timedelta
df['ride_length'] = pd.to_timedelta(df['ride_length'])

# Convert to minutes
df['ride_length_minutes'] = df['ride_length'].dt.total_seconds() / 60
```

```python
df['hour'] = df['started_at'].dt.hour
df['month'] = df['started_at'].dt.month_name()
df['day_of_week'] = df['started_at'].dt.day_name()
```

```python
avg_duration = df.groupby('member_casual')['ride_length_minutes'].mean()
print(avg_duration)
```

```python
avg_duration = df.groupby('member_casual')['ride_length_minutes'].mean().round(2)
print("Average Ride Duration (minutes):")
print(avg_duration)
```

```python
rides_by_day = df.groupby(['member_casual', 'day_of_week']).size().unstack()
print("\nRide Counts by Day of Week:")
print(rides_by_day)
```

```python
bike_type_counts = df.groupby(['member_casual', 'rideable_type']).size().unstack()
print("\nBike Type Usage:")
print(bike_type_counts)
```

```python
rides_by_month = df.groupby(['member_casual', 'month']).size().unstack()
print("\nRides by Month:")
print(rides_by_month)
```
